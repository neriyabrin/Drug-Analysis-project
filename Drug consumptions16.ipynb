```python

def percent(data,ce,category_name,category_sum):
    
    category=list(data[ce])
    count=list(data['sum'])
    perc=[]
    i=0
    j=0
    while( i < len(category)):
        check=0
        while(category[i]==category_name[j]):
            perc.append((count[i]/category_sum[j])*100)
            check=1
            j+=1
            if(j==len(category_name)):
                j=0    
        if(check==0):
            j+=1
        else:
            i+=1
        if(j>len(category_name)-1):
            j=0
    new_data=data.assign(Percentage = perc)
    return new_data
```

```python
age1=new_file[new_file.loc[:,['Age']]=='18-24']['Age'].count()
age2=new_file[new_file.loc[:,['Age']]=='25-34']['Age'].count()
age3=new_file[new_file.loc[:,['Age']]=='35-44']['Age'].count()
age4=new_file[new_file.loc[:,['Age']]=='45-54']['Age'].count()
age5=new_file[new_file.loc[:,['Age']]=='55-64']['Age'].count()
age6=new_file[new_file.loc[:,['Age']]=='65+']['Age'].count()

age_name=["18-24","25-34","35-44","45-54","55-64","65+"]
age_sum=[age1,age2,age3,age4,age5,age6]

B=new_file.groupby('Alcohol')[['Age']].value_counts().reset_index()
B=B.rename(index=str,columns={0:'sum'})
A_age=percent(B,'Age',age_name,age_sum)

B=new_file.groupby('Cannabis')[['Age']].value_counts().reset_index()
B=B.rename(index=str,columns={0:'sum'})
C_age=percent(B,'Age',age_name,age_sum)
```

```python
def with_hue(ax, feature, Number_of_categories, hue_categories):
    a = [p.get_height() for p in ax.patches]
    patch = [p for p in ax.patches]
    for i in range(Number_of_categories):
        for j in range(hue_categories):
            percentage = '{:.0f}%'.format(100 * a[(j*Number_of_categories + i)]/100)
            
            x = patch[(j*Number_of_categories + i)].get_x() + patch[(j*Number_of_categories + i)].get_width() / 2 
            
            y = patch[(j*Number_of_categories + i)].get_y() + patch[(j*Number_of_categories + i)].get_height() + 1.5
           
            ax.annotate(percentage, (x, y), size = 8,ha='center')
```

```python
fig, axes = plt.subplots(figsize=(20,8), ncols=2)
alcohol=sns.barplot(data=A_age,ax=axes[0],x='Age',y='Percentage',hue="Alcohol",order=age_name,palette= "Set1")
cannabis=sns.barplot(data=C_age,ax=axes[1],x='Age',y='Percentage',hue="Cannabis",order=age_name,palette= "Set1")
axes[0].set_ylim(0,80)
axes[1].set_ylim(0,80)
with_hue(axes[0],A_age.Alcohol,7,6)
with_hue(axes[1],C_age.Cannabis,7,6)
plt.show()
alcohol.set_title('Alcohol by Age', fontsize = 15 , color = 'black',fontstyle = 'italic' , backgroundcolor='blue')
cannabis.set_title('Cannabis by Age', fontsize = 15 , color = 'black',fontstyle = 'italic' , backgroundcolor='blue')

plt.show()
```
